```python
import os
os.environ["PATH"] += os.pathsep + 'D:/Program Files (x86)/Graphviz/bin/'
import pandas as pd
from sklearn import tree
```

#Read data

```python
#Read data
df_mx = pd.read_csv("Sample Data/MX.csv", index_col=0)
df_thai = pd.read_csv("Sample Data/THAI.csv", index_col=0)
df_sgp = pd.read_csv("Sample Data/SGP.csv", index_col=0)

df_agg = pd.concat([df_mx,df_sgp, df_thai])
#df_test = pd.concat([df_sgp])
df_agg.head()
```

#Explore the Data

#Classification with Scikit-Learn

```python
from sklearn import svm
import numpy as np

# è®€å…¥è³‡æ–™
X = df_agg.drop(columns=['Target'])
X = X.to_numpy()
y = df_agg['Target'].values
X
```

```python
clf = tree.DecisionTreeClassifier()
clf = clf.fit(X, y)
```

```python
tree.plot_tree(clf)
```

```python
import graphviz 
dot_data = tree.export_graphviz(clf, out_file=None) 
graph = graphviz.Source(dot_data) 
#graph.render("CCSP") 
```

```python
# the array must be change into some format to show the names
dot_data = tree.export_graphviz(clf, out_file=None, 
                     #feature_names=X.feature_names,  
                     #class_names=y.target_names,  
                     filled=True, rounded=True,  
                     special_characters=True)  
graph = graphviz.Source(dot_data)  
graph 
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.tree import export_text

decision_tree = DecisionTreeClassifier(random_state=0, max_depth=50)
decision_tree = decision_tree.fit(X, y)
r = export_text(decision_tree)
print(r)
```
