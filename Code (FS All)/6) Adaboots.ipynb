```python
# Load libraries
import pandas as pd
from sklearn.ensemble import AdaBoostClassifier
from sklearn import datasets
# Import train_test_split function
from sklearn.model_selection import train_test_split
#Import scikit-learn metrics module for accuracy calculation
from sklearn import metrics

#score
from sklearn.metrics import precision_score, accuracy_score, roc_auc_score, confusion_matrix,f1_score, roc_curve, auc, recall_score
```

```python
#Read data
df_agg = pd.read_csv("Sample Data/FS_ALL.csv", index_col=0)
df_agg.head()
```

```python
X = df_agg.drop(columns=['Target'])
y = df_agg['Target']
```

```python
# Split dataset into training set and test set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3) # 70% training and 30% test
```

```python
# Create adaboost classifer object
abc = AdaBoostClassifier(n_estimators=50,
                         learning_rate=1)
# Train Adaboost Classifer
model = abc.fit(X_train, y_train)

#Predict the response for test dataset
y_pred = model.predict(X_test)
```

```python
# Model Accuracy, how often is the classifier correct?
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```

```python
# Load libraries
from sklearn.ensemble import AdaBoostClassifier

# Import Support Vector Classifier
from sklearn.svm import SVC
#Import scikit-learn metrics module for accuracy calculation
from sklearn import metrics
svc=SVC(probability=True, kernel='linear')

# Create adaboost classifer object
abc =AdaBoostClassifier(n_estimators=50, base_estimator=svc,learning_rate=1)

# Train Adaboost Classifer
model = abc.fit(X_train, y_train)

#Predict the response for test dataset
y_pred = model.predict(X_test)


# Model Accuracy, how often is the classifier correct?
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```

```python
# ç¸¾æ•ˆ


print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
print("Accuracy:", accuracy_score(y_test, y_pred))

print("Precision Score",precision_score(y_test, y_pred, average='macro'))
print("Precision Score (Micro", precision_score(y_test, y_pred, average='micro'))
print("Precision Score (Weighted)",precision_score(y_test, y_pred, average='weighted'))
print("Precision Score", precision_score(y_test, y_pred, average=None))

print("Recall Score (Macro)", recall_score(y_test, y_pred, average='macro',zero_division=1))
print("Recall Score (Micro)", recall_score(y_test, y_pred, average='micro',zero_division=1))
print("Recall Score (Weight)",recall_score(y_test, y_pred, average='weighted',zero_division=1))
print("Recall Score (Avg None)",recall_score(y_test, y_pred, average=None,zero_division=1))


#print("Roc_Score:", roc_auc_score(y, clf.decision_function(X)))      
print("F1 Score (Macro):", f1_score(y_test, y_pred, average='macro'))
print("F1 Score (Micro):", f1_score(y_test, y_pred, average='micro'))
print("F1 Score (Weighted):", f1_score(y_test, y_pred, average='weighted'))
print("F1 Score (Avg None):", f1_score(y_test, y_pred, average=None))
print("F1 Score:", f1_score(y_test, y_pred, zero_division=1))
```
